```python
import wget
import pandas as pd
import zipfile
import glob
```

```python
years = range(2007,2021)
types = ["Accidents","Drivers", "Pedestrians", "Occupants", "Vehicles"]
```

```python
years = range(2001,2021)
accidents = []
cols = []
for y in years:
    try:
        data = pd.read_csv(f"data/{y}Accidents.csv")
        print(data.shape)
        for c in data.columns:
            if c not in cols:
                cols.append(c)
        accidents.append(data.loc[data['municipality_name'] == 'JERSEY CITY'])
    except:
        pass
accidents = pd.concat(accidents)
```

```python
years = range(2001,2021)
pedestrians = []
cols = []
for y in years:
    data = pd.read_csv(f"data/{y}Pedestrians.csv")
    print(data.shape)
    for c in data.columns:
        if c not in cols:
            cols.append(c)
    pedestrians.append(data.loc[data['is_bycyclist'] == "Y"])
pedestrians = pd.concat(pedestrians)
```

```python
accidents.describe()
```

```python
bike_accidents = accidents.merge(pedestrians, on="id")
bike_accidents = bike_accidents[['id', 'county_name', 'municipality_name', 'crash_date','crash_location', 'cross_street_name',  'latitude', 'longitude']]
bike_accidents.columns = ['id', 'countyname', 'municipalityname', 'crashdate','crashlocation', 'crossstreetname',  'latitude', 'longitude']
bike_accidents['state'] = "NJ"
bike_accidents["longitude"] = -bike_accidents["longitude"]
```
